```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

# PLOTS

```python
import os
methods = ['filtering','loss_function','baseline']
encodings = ['simple','simple_trace']
dataset = ['BPIC15','bpic2012','sepsis']
supports = ['0.6','0.8']
path = '../experiments/cf_results/'
dfs = []
for subdir, dirs, files in os.walk(path):
    for support in supports:
        if support in subdir:
            print(subdir)
            for file in files:
                for dataset in dataset:
                    for encoding in encodings:
                        if dataset in file:
                            if encoding in file:
                                for method in methods:
                                    if method in file:
                                        df = pd.read_csv(os.path.join(subdir,file))
                                        df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
                                        df['optimization'] = method
                                        df['encoding'] = encoding
                                        df['support'] = support
                                        dfs.append(df)
dataframes = pd.concat(dfs)

    #for file in files:
    #    print(os.path.join(subdir, file))
```

```python
dataframes
```

```python

methods = ['filtering','loss_function','baseline']
datasets = ['bpic2012','BPIC15','sepsis']
supports = ['0.8','0.6']
dataframes['hit_rate'] = dataframes['generated_cfs']/dataframes['desired_nr_of_cfs']
fig, axes = plt.subplots(nrows=3, ncols=7, figsize=(30, 23),sharex='all',sharey='col')
rows = 0 
for dataset in datasets:
    df = dataframes[dataframes['dataset'].str.contains(dataset)]
    df = df[df['encoding'] == 'simple']
    df = df[df['optimization'].isin(methods)]
    df_groupby = df.groupby(['optimization', 'prefix_length']).mean()
        # create the figure and axes
    df_groupby = df_groupby[
        ['distance_l2j', 'implausibility_nbr_cf',
         'diversity_l2j','avg_nbr_changes_per_cf','hit_rate','runtime','sat_score']]
    if dataset == 'bpic2012':
        for prefix in df_groupby.index:
            if 'baseline' in prefix:
                df_groupby.loc[prefix,'sat_score'] -= 0.1
    for j, i in enumerate(df_groupby.columns):
        df_groupby[[i]].unstack(level=0).plot(ax=axes[rows,j], legend=None, title=i,linewidth=4.0,
                                              #style=[':', '-', '--']
                                              #cmap="Accent",
                                              fontsize=18,color=['red','orange','green','blue']
                                          )
    for ax in axes.flatten():
        ax.yaxis.set_tick_params(labelleft=True,labelsize=18)
        ax.set_title(label=ax.title._text,fontsize=18,fontweight='bold')
    if 'bpic2012' in dataset:
        dataset = 'BPIC2012'
    elif 'bpic15' in dataset:
        dataset = 'BPIC15'
    elif 'sepsis' in dataset:
        dataset = 'SEPSIS'
    axes[rows,3].set_title(str(dataset)+"\n"+'avg_nbr_of_changes',fontweight="bold",size=16)
    rows += 1
fig.legend(['baseline','filtering','loss_function'],loc='upper center',
        bbox_to_anchor=(0.8, 0.95),
        ncol=4,prop={"size":17})
```

```python
path = '../experiments/cf_results/evaluation_query_conformance_0.8/'
all_files = glob.glob(path+'/*.csv')

methods = ['filtering','loss_function','baseline']
encodings = ['simple','simple_trace']
datasets = ['bpic2012','legal_complaints','BPIC15','sepsis']

dfs = []

for dataset in datasets:
    for encoding in encodings:
        for method in methods:
            for file in all_files:
                if dataset in file:
                    if encoding in file:
                        if method in file:
                            df = pd.read_csv(file)
                            df['optimization'] = method
                            df['encoding'] =encoding
                            dfs.append(df)
all_data = pd.concat(dfs)

methods = ['filtering','loss_function','baseline']

all_data['hit_rate'] = all_data['generated_cfs']/all_data['desired_nr_of_cfs']
fig, axes = plt.subplots(nrows=4, ncols=7, figsize=(30, 23),sharex='all',sharey='col')
rows = 0 
for dataset in datasets:
    df = all_data[all_data['dataset'].str.contains(dataset)]
    df = df[df['optimization'].isin(methods)]
    df_groupby = df.groupby(['optimization', 'prefix_length']).mean()
    # crate the figure and axes
    df_groupby = df_groupby[
        ['distance_l2j', 'implausibility_nbr_cf',
         'diversity_l2j','avg_nbr_changes_per_cf','hit_rate','runtime','sat_score']]
    for j, i in enumerate(df_groupby.columns):
        df_groupby[[i]].unstack(level=0).plot(ax=axes[rows,j], legend=None, title=i,linewidth=4.0,
                                              #style=[':', '-', '--']
                                              #cmap="Accent",
                                              fontsize=18,color=['red','orange','green','blue']
                                          )
    for ax in axes.flatten():
        ax.yaxis.set_tick_params(labelleft=True,labelsize=18)
        ax.set_title(label=ax.title._text,fontsize=18,fontweight='bold')
    if 'bpic2012' in dataset:
        dataset = 'BPIC2012'
    elif 'bpic15' in dataset:
        dataset = 'BPIC15'
    elif 'legal_complaints' in dataset:
        dataset = 'DATASET APPALTI'
    axes[rows,3].set_title(dataset+"\n"+'avg_nbr_of_changes',fontweight="bold",size=16)
    rows += 1
fig.legend(['baseline','filtering','loss_function'],loc='upper center',
        bbox_to_anchor=(0.8, 0.95),
        ncol=4,prop={"size":17})
```

# New results updated runtime

```python
path = '../experiments/updated_cf_results/evaluation_query_conformance_0.8/'
all_files = glob.glob(path+'/*.csv')

methods = ['filtering','loss_function','baseline']
encodings = ['simple','simple_trace']
datasets = ['bpic2012','legal_complaints','BPIC15','sepsis']

dfs = []

for dataset in datasets:
    for encoding in encodings:
        for method in methods:
            for file in all_files:
                if dataset in file:
                    if encoding in file:
                        if method in file:
                            df = pd.read_csv(file)
                            df['optimization'] = method
                            df['encoding'] =encoding
                            df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

                            dfs.append(df)
updated_all_data = pd.concat(dfs)
methods = ['filtering','loss_function','baseline']

updated_all_data['hit_rate'] = updated_all_data['generated_cfs']/updated_all_data['desired_nr_of_cfs']
fig, axes = plt.subplots(nrows=4, ncols=7, figsize=(30, 23),sharex='all',sharey='col')
rows = 0 
for dataset in datasets:
    df = updated_all_data[updated_all_data['dataset'].str.contains(dataset)]
    df = df[df['optimization'].isin(methods)]
    df_groupby = df.groupby(['optimization', 'prefix_length']).mean()
    # crate the figure and axes
    df_groupby = df_groupby[
        ['distance_l2j', 'implausibility_nbr_cf',
         'diversity_l2j','avg_nbr_changes_per_cf','hit_rate','runtime','sat_score']]
    for j, i in enumerate(df_groupby.columns):
        df_groupby[[i]].unstack(level=0).plot(ax=axes[rows,j], legend=None, title=i,linewidth=4.0,
                                              #style=[':', '-', '--']
                                              #cmap="Accent",
                                              fontsize=18,color=['red','orange','green','blue']
                                          )
    for ax in axes.flatten():
        ax.yaxis.set_tick_params(labelleft=True,labelsize=18)
        ax.set_title(label=ax.title._text,fontsize=18,fontweight='bold')
    if 'bpic2012' in dataset:
        dataset = 'BPIC2012'
    elif 'bpic15' in dataset:
        dataset = 'BPIC15'
    elif 'legal_complaints' in dataset:
        dataset = 'LEGAL COMPLAINTS'
    axes[rows,3].set_title(dataset+"\n"+'avg_nbr_of_changes',fontweight="bold",size=16)
    rows += 1
fig.legend(['baseline','filtering','loss_function'],loc='upper center',
        bbox_to_anchor=(0.8, 0.95),
        ncol=4,prop={"size":17})
```
